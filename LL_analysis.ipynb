```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the clean LL file

dframe = pd.read_csv("ll_cleaned.csv", na_values=' ')
dframe
```

```python
# Display the dataframe info()

dframe.info()
```

```python
# plotting the number of Liquor License in each year

fig,ax = plt.subplots(figsize=(8,5))
sns.countplot(dframe['LicenseYear'])
```

```python
# plotting the number of Liquor License in each zip code

fig,ax = plt.subplots(figsize=(16,5))
sns.countplot(dframe['AddrZip'])
```

```python
# Count of License Status each Year

plt.figure(figsize=(15,6))
sns.countplot(x = "LicenseYear", hue = "LicenseStatus", data = dframe)
plt.title("Count of License Status each Year")
plt.show()
```
